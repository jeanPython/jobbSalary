```python
jobRatingId =[]
id=0
for jobToUpdate in job['Rating']:
    jobRatingId.append({'id':id,'Rating':jobToUpdate})
    id+=1
    
for index in job.index:
    for transform in jobRatingId:
        if transform['Rating'] == job['Rating'][index]:
            job['Rating'][index] = ''
            job['Rating'][index] =transform['id']
```

```python
jobTitleId =[]
id =0
for jontitles in job['Job Title'].unique().tolist():
    jobTitleId.append({'id':id,'tittle':jontitles})
    id +=1

#vamos a transformar los datos para cada una de las industrias
for index in job.index:
    #remove the glassdoor text
    job['Salary Estimate'][index] = job['Salary Estimate'][index].replace(' (Glassdoor est.)','')
    for transform in jobTitleId:
        #print(jobTitleId[transform])
        if transform['tittle'] == job['Job Title'][index]:
            job['Job Title'][index] = transform['id']    
```

```python
jobSalaryId =[]
id=0
for jobToUpdate in job['Salary Estimate'].unique().tolist():
    jobSalaryId.append({'id':id,'salary':jobToUpdate})
    id +=1

for index in job.index:
    for transform in jobSalaryId:
        if transform['salary'] == job['Salary Estimate'][index]:
            job['Salary Estimate'][index] = transform['id']
```
